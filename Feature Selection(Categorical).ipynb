```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df=pd.read_csv('banana_quality_dataset.csv')
```

```python
df
```

```python
Y=df['quality_category']
```

```python
X=df.drop(['sample_id','quality_category','harvest_date'],axis=1)
```

```python
X
```

```python
df['variety'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder,MinMaxScaler
from sklearn .compose import ColumnTransformer
```

```python
transformer =ColumnTransformer(transformers=[("t1",OrdinalEncoder(),[0,1,4,]),
                                             ("t2",MinMaxScaler(),[2,3,5,6,7,8,9,10,11,12])])
X_transformed=transformer.fit_transform(X)
```

```python
X_transformed=pd.DataFrame(X_transformed,columns=X.columns)
```

```python
X_transformed
```

```python
X_transformed.head()
```

# Mutual Information for Classification

```python
from sklearn.feature_selection import mutual_info_classif

mutual_info=mutual_info_classif(X_transformed,Y)
mutual_info=pd.DataFrame({
    'Feature':X_transformed.columns,
    'Mi_Scores':mutual_info
}).sort_values(by='Mi_Scores',ascending=False)
```

```python
mutual_info
```

# Feature Selection Using CHi2 Test

```python
from sklearn.feature_selection import chi2

# Apply Chi -Square Test
chi_scores,_=chi2(X_transformed,Y)

# Create DataFrameS for only chiSquare Scores

chi_df=pd.DataFrame({
    'Feature':X_transformed.columns,
    'Chi2_Score':chi_scores
}).sort_values(by='Chi2_Score',ascending=False)
```

```python
chi_df
```

# SelectKbest

```python
from sklearn.feature_selection import chi2,SelectKBest
select_columns=SelectKBest(chi2,k=5)
best_features=select_columns.fit_transform(X_transformed,Y)
```

```python
selected_features=X.columns[select_columns.get_support()]
```
